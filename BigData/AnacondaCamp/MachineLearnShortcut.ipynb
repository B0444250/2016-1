The scikit-learn package is built of many sub-packages that each cover a different part of machine learning. You will almost always interact with scikit-learn by importing a specific class or function in the following manner.

### Regressor vs Classifier
Notice that most of the supervised machine learning methods have one class each for regression and classification.

### Sklearn gotchas
+ By convention put your input array into a variable named **`X`** and target array into variable **`y`**.
+ Convert both input and output to numpy arrays before using scikit-learn
+ Your input array must be two-dimensional, even if there is only one predictor variable
+ Although, you can have multiple labels for your **`y`** for each observation, this is very rare. When you have a single label for each observation, convert your **`y`** into a one dimensional numpy array.
+ **`X`** and **`y`** must not contain any missing values
+ **`X`** must be all numeric.
+ **`y`** can be an array of strings or numbers

### Typical workflow from pandas to prediction
Here is an outline of a very basic workflow for going from initial data pull to machine learning
> `read data in pandas dataframe` 
> `lots of pandas manipulations`  
> `Do lots of EDA`  

> `Still in pandas, prepare for machine learning`  
> `Remove missing values`  
> `Binarize columns with string values with pd.get_dummies`  

> `move to sklearn by converting pandas objects to numpy arrays X and y`  
> `Scale continuous features`  
> `Other transformation to either features or target variables - take log, etc...`  

> `Split data - kfold, or validation set`  
> `Import estimator from sklearn`  
> `Instantiate estimator`  
> `Use fit method`  
> `Use predict or score method to validate model`

Classification: Predict Categorical Data
===
</code>
Predict the class, or label $(t)$, of a sample based on its features $(x)$. Examples: Recognize hand-written digits, or mark email as spam. In scikit-learn, labels are represented as integers and get expanded internally into matrices of binary choices between unique integer labels. Use class_weight='balanced' in most models to adjust for unbalanced datasets (more training data from one class than others). Training data has $N$ samples and $D$ features.

  -  Logistic Regression: X, input, Y, output in value 0 or 1. $p,\color{red}{1-p}={\cal P}(Y=1|X),\color{red}{{\cal P}(Y=0|X)}$, and
     
     \begin{eqnarray}
        \ln\left(\frac{p}{1-p}\right)&=&w_0+w_1x_1+w_nx_n+\epsilon &(=g(x_i))\\
        p&=& \frac{1}{1+e^{g(x_i)}}
     \end{eqnarray}
     which maps data from $\mathbb R$ to $(0,1)$. <img src="ML/C1.png" width=70%></img>

```
      s_df.loc[s_df.isnull()] = 0  Use mask to replace NaN

      s_df.fillna(method='ffill')  Fill forward (last valid value)

```   

 

<code style="background-color:darkblue;color:white;font-size:2em;font-face:bold">
Reshaping Dataframes/Pivot Tables
---
</code>

 Tools for reshaping DataFrames from the wide to the long format and back. The long format can be tidy, which means that "each variable is a column, each observation is a row"1. Tidy data is easier to filter, aggregate, transform, sort, and pivot.  Reshaping operations often produce multi-level indices or columns, which can be sliced and indexed.
 


**MultiIndex**: A Multi-Level Hierarchical Index

Often created as a result of:
```
  > df.groupby(list_of_columns)
  > df.set_index(list_of_columns)
```
Contiguous labels are displayed together but apply to each row. The concept is similar to multi-level columns.

A **MultiIndex** allows indexing and slicing one or multiple levels at once. Using the Long example from the right:
```
                        long.loc[1900]  Pivot Tables
            long.loc[(1900, 'March')]   value2
      long.xs('March', level='Month')   All March rows
```      
Simpler than using boolean indexing, for example:
```
> long[long.Month == 'March']
```



**Long to Wide Format and Back with stack() and unstack()**



Pivot column level to index, i.e. "stacking the columns" (wide to long):
```
 > df.stack()
```
Pivot index level to columns, "unstack the columns" (long to wide):
```
 > df.unstack()
```
If multiple indices or column levels, use level number or name to stack/unstack:
```
 > df.unstack(0) or > df.unstack('Year')
```
A common use case for unstacking, plotting group data vs index after groupby:
```
 > (df.groupby(['A', 'B])['relevant'].mean()
      .unstack().plot())
```
